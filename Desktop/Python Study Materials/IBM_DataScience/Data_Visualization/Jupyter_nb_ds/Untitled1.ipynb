```python
df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], inplace=True, axis=1)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

## Data Visualisation using matplotlib

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
india_dataset = df_can.loc['India', years]
india_dataset.plot(kind='line')
plt.title('Immigration from India')
plt.xlabel('Years')
plt.ylabel('No. of immigrants')
plt.show()
```

#### Plot comparing two values

```python
df_can_1 = df_can.loc[['United Kingdom of Great Britain and Northern Ireland', 'Germany'], years]
df_can_2 = df_can_1.transpose()
df_can_2.head()
```

```python
df_can_2.plot(kind='line')
```

```python
df_can.sort_values(by='Total', ascending=False, inplace=True, axis=0)
df_can_sort = df_can.iloc[0:5, 3:-1]
df_can_transpose = df_can_sort.transpose()

df_can_transpose.plot(kind='line', figsize=(15,10))
plt.title('Top 5 Immigrant Country to Canada')
plt.xlabel('Year')
plt.ylabel('No. of immigrants')
plt.show()
```
