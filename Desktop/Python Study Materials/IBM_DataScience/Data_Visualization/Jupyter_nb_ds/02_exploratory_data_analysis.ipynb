## P Value
### The P-value is the probability value that the correlation between these two variables is statistically significant

```python
from scipy import stats
```

```python
pearson_coeff, p_val = stats.pearsonr(df['wheel-base'], df['price'])
```

```python
print('Pearson coefficient is ', pearson_coeff, ' p value ', p_val)
```

#### PC is close to 0.5 which means the linear relationship is not strong enough, but p value < 0.001 which means the corelation is statistically significant

```python
p_c, p_v = stats.pearsonr(df['horsepower'], df['price'])
print('Pearson coefficient is ', p_c, ' p value ', p_v)
```

#### PC is close to 1 which means there is a positive linear relationship; p value < 0.001 which eans the correlation is statistically significant

```python
p_c1, p_v1 = stats.pearsonr(df['length'], df['price'])
print('Pearson coefficient is ', p_c1, ' p value ', p_v1)
```

#### PC is 0.6 which means there is a weak linear relationship, but low P value indicate correlation is staitically significant

```python
p_c2, p_v2 = stats.pearsonr(df['curb-weight'], df['price'])
print('Pearson coefficient is ', p_c2, ' p value ', p_v2)
```

#### PC is close to 1 which indicate a strong linear relationship and very low p value means the correlation is statistically significant
